```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
x,y_true = make_moons(n_samples = 500, noise=0.05, random_state=42)
```

```python
df = pd.DataFrame(x, columns=['Feature_1','Feature_2'])
```

```python
df
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(df)
```

```python
final_kmean = KMeans(n_clusters=2, random_state=42)
cluster_labels = final_kmean.fit_predict(x_scaled)
```

```python
df['cluster'] = cluster_labels
```

```python
sns.scatterplot(x='Feature_1', y='Feature_2', hue='cluster', data=df)
plt.title('Clusters')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(x_scaled)
```

```python
df['dbscan_cluster'] = dbscan_labels
```

```python
sns.scatterplot(x='Feature_1', y='Feature_2', hue='dbscan_cluster', data=df)
plt.title('DBSCAN Clusters')
plt.show()
```
